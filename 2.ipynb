```python
df.columns = df.columns.str.replace(" ", "")
df.columns
```

```python
#Changing it dummy variable and dropping them

df_dummies = pd.get_dummies(df[['workclass','education','marital-status','occupation','relationship','race','sex','native-country','income']])
df_dummies
df_dummies.dtypes

dfk = df.join(df_dummies)
dfk = dfk.drop(['workclass','education','marital-status','occupation','relationship','race','sex','native-country','income'], axis=1)
dfk = dfk.drop(['workclass_ Federal-gov','education_ 10th','marital-status_ Divorced','occupation_ Adm-clerical','relationship_ Husband','race_ Amer-Indian-Eskimo','sex_ Female','native-country_ Cambodia'],1)

display(HTML(dfk.head(5).to_html()))
```

```python
# build linear regression models by using hold-out evaluations ###################################
# there are different ways: https://www.freecodecamp.org/news/data-science-with-python-8-ways-to-do-linear-regression-and-measure-their-speed-b5577d75f8b/
# unlike R, Python libraries do not provide feature selection for regression models
# but you can still find some self-implemented forward/backward selection codes online, if you need it

from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
import sklearn.metrics as metrics
from sklearn.model_selection import train_test_split
```

```python
# using hold-out evaluation for example
x_train, x_test, y_train, y_test = train_test_split(dfk, dfk['age'], test_size=0.3)

print('train set:\n',x_train.shape, "\n", y_train.shape)
print('test set:\n',x_test.shape, "\n", y_test.shape)
```

```python
# method 1: using the LinearRegression model
# API: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html
# drawback: you can get limited information about the model
m1=LinearRegression()
m1.fit(x_train, y_train)
print("\nBuild mdoels by LienarRegression()")
print("Intercept: ", m1.intercept_,"\nCoefficients:\n", m1.coef_,"\nR-square: ", m1.score(x_train, y_train))
```

```python
# method 2: using OLS in statsmodels
# API: https://www.statsmodels.org/stable/generated/statsmodels.regression.linear_model.OLS.html
print("\nBuild mdoels by OLS()")
m2 = sm.OLS(y_train, x_train)
rst=m2.fit()
print("R-square: ", rst.rsquared, "\n\nCoefficients:\n", rst.params, "\n\np-values: \n", rst.pvalues)
print(rst.summary())
```

```python
# evaluate models by MAE or RMSE
y_predict=rst.predict(x_test)
MAE=metrics.mean_absolute_error(y_test, y_predict)
RMSE=np.sqrt(metrics.mean_squared_error(y_test, y_predict))
print("\nEvaluations on the test set: \nMAE: ", MAE,", RMSE: ", RMSE)
```

```python
# build linear regression models by using N-fold cross evaluations ###################################

from sklearn.model_selection import cross_val_score
from sklearn.base import BaseEstimator, RegressorMixin

m=LinearRegression()
rst=cross_val_score(m, x_train, y_train, cv=5, scoring='neg_mean_squared_error')
mse = np.abs(rst.mean())
print("\n5-fold Evaluations for LinearRegression(): \n RMSE: ", np.sqrt(mse))


class SMWrapper(BaseEstimator, RegressorMixin):
    """ A universal sklearn-style wrapper for statsmodels regressors """
    def __init__(self, model_class, fit_intercept=True):
        self.model_class = model_class
        self.fit_intercept = fit_intercept
    def fit(self, X, y):
        if self.fit_intercept:
            X = sm.add_constant(X)
        self.model_ = self.model_class(y, X)
        self.results_ = self.model_.fit()
    def predict(self, X):
        if self.fit_intercept:
            X = sm.add_constant(X)
        return self.results_.predict(X)
    

data_x = sm.add_constant(dfk.drop('age',1))
data_y = dfk['age']
RMSE = np.sqrt(-cross_val_score(SMWrapper(sm.OLS), data_x, data_y, scoring="neg_mean_squared_error", cv = 10))
print("RMSE in 10 folds: ", RMSE)
print("RMSE = ", np.mean(RMSE))
```

```python
# 02. Regression with Regularizations #############################
from sklearn.linear_model import Lasso
from sklearn.linear_model import LassoCV
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import ElasticNetCV
from sklearn.metrics import mean_squared_error


# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Lasso.html#sklearn.linear_model.Lasso
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LassoCV.html#sklearn.linear_model.LassoCV
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html#sklearn.linear_model.Ridge
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.RidgeCV.html#sklearn.linear_model.RidgeCV
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.ElasticNet.html#sklearn.linear_model.ElasticNet
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.ElasticNetCV.html#sklearn.linear_model.ElasticNetCV

m1=Lasso(alpha=0.1)
m1.fit(x_train, y_train)
print("\nBuild mdoels by Lasso()")
print("Intercept: ", m1.intercept_,"\nCoefficients:\n", m1.coef_,"\nR-square: ", m1.score(x_train, y_train))
```

```python
# evaluate models by MAE or RMSE
y_predict=m1.predict(x_test)
MAE=metrics.mean_absolute_error(y_test, y_predict)
RMSE=np.sqrt(metrics.mean_squared_error(y_test, y_predict))
print("\nEvaluations on the test set: \nMAE: ", MAE,", RMSE: ", RMSE)
# N-fold cross validation
x = dfk.drop('age',1)
y = dfk['age'] 
m1=LassoCV(cv=5, random_state=0).fit(x, y)
print("\nBuild mdoels by LassoCV()")
print("Intercept: ", m1.intercept_,"\nCoefficients:\n", m1.coef_,"\nR-square: ", m1.score(x,y))
mse=mean_squared_error(y, m1.predict(x))
print("RMSE = ", np.sqrt(mse))

# other methods are similar to the Lasso and LassoCV above
# double check the API to learn how to change the parameters
```

```python
# 03. Adapt Classifications to Numerical Predictions ###############################
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR

```

```python
# KNN, https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html
m1=KNeighborsRegressor(n_neighbors=5)
m1.fit(x_train, y_train)
print("\nBuild mdoels by KNN")
# evaluate models by MAE or RMSE
y_predict=m1.predict(x_test)
MAE=metrics.mean_absolute_error(y_test, y_predict)
RMSE=np.sqrt(metrics.mean_squared_error(y_test, y_predict))
print("\nEvaluations on the test set: \nRMSE: ", RMSE)
# N-fold cross validation
m=KNeighborsRegressor(n_neighbors=5)
rst=cross_val_score(m, x, y, cv=5, scoring='neg_mean_squared_error')
mse = np.abs(rst.mean())
print("\nEvaluations: \n RMSE: ", np.sqrt(mse))


```

```python
# Neural Networks, https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPRegressor.html#sklearn.neural_network.MLPRegressor
m1=MLPRegressor(random_state=1, max_iter=200)
m1.fit(x_train, y_train)
print("\nBuild mdoels by Neural Networks")
# evaluate models by MAE or RMSE
y_predict=m1.predict(x_test)
MAE=metrics.mean_absolute_error(y_test, y_predict)
RMSE=np.sqrt(metrics.mean_squared_error(y_test, y_predict))
print("\nEvaluations on the test set: \nRMSE: ", RMSE)
# N-fold cross validation
m=MLPRegressor(random_state=1, max_iter=200)
rst=cross_val_score(m, x, y, cv=5, scoring='neg_mean_squared_error')
mse = np.abs(rst.mean())
print("\nEvaluations: \n RMSE: ", np.sqrt(mse))


```

```python
# Regression Tree, https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
m1=DecisionTreeRegressor(random_state=0)
m1.fit(x_train, y_train)
print("\nBuild mdoels by Regression Tree")
# evaluate models by MAE or RMSE
y_predict=m1.predict(x_test)
MAE=metrics.mean_absolute_error(y_test, y_predict)
RMSE=np.sqrt(metrics.mean_squared_error(y_test, y_predict))
print("\nEvaluations on the test set: \nRMSE: ", RMSE)
# N-fold cross validation
m=DecisionTreeRegressor(random_state=0)
rst=cross_val_score(m, x, y, cv=5, scoring='neg_mean_squared_error')
mse = np.abs(rst.mean())
print("\nEvaluations: \n RMSE: ", np.sqrt(mse))


```
